```python
# logistica
import matplotlib.pyplot as fig

r = 2.9
a =0.1
x=[a]

for k in range(100):
  a = r*a*(1-a)
  x.append(a)

fig.plot(x, '-k',linewidth=2)
fig.xlabel('Itera√ßao',fontsize=16)
fig.ylabel('X[k]',fontsize=24)
fig.title(f'Logistica com r = {r}',fontsize=16)
# fig.show()
```
